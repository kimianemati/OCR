```python
import easyocr
import cv2
import numpy as np


reader = easyocr.Reader(['fa'])


image_path = '7.jpg'
image = cv2.imread(image_path)


height, width = image.shape[:2]
new_width = 2 * width
new_height = 2 * height
resized_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)


gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)


alpha = 1.5
beta = 30
adjusted_image = cv2.convertScaleAbs(gray_image, alpha=alpha, beta=beta)


kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
sharpened_image = cv2.filter2D(adjusted_image, -1, kernel)




blurred_image = cv2.GaussianBlur(sharpened_image, (5, 5), 0)


_, thresholded_image = cv2.threshold(blurred_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)


cv2.imwrite('processed_image.jpg', thresholded_image)




results = reader.readtext('processed_image.jpg')


confidence_threshold = 0.5

for result in results:
    detected_text = result[1]
    confidence = result[2]
    if confidence >= confidence_threshold:
        print(f'Detected text: {detected_text}, Confidence: {confidence:.2f}')
    else:
        print(f"Low confidence in detected text: {detected_text}, Confidence: {confidence:.2f}")
```
